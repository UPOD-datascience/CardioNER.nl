```python
import pandas as pd
import os
import pyarrow as pa
import pyarrow.parquet as pq

from ftplib import FTP

from collections import defaultdict

from tqdm import tqdm

from list_pmc_files import list_pmc_files_detailed, load_and_concatenate_csvs
```

```python
ftp_comm_remote = 'https://ftp.ncbi.nlm.nih.gov/pub/pmc/oa_bulk/oa_comm/xml/'
ftp_noncomm_remote = 'https://ftp.ncbi.nlm.nih.gov/pub/pmc/oa_bulk/oa_noncomm/xml/'
ftp_misc_remote =  'https://ftp.ncbi.nlm.nih.gov/pub/pmc/oa_bulk/oa_other/xml/'

contains_file = 'filelist.csv'


comm_csv = list_pmc_files_detailed(ftp_comm_remote, filename_pattern=contains_file)
noncomm_csv = list_pmc_files_detailed(ftp_noncomm_remote, filename_pattern=contains_file)
misc_csv = list_pmc_files_detailed(ftp_misc_remote, filename_pattern=contains_file)

contains_file = '.tar.gz'

comm_xml = list_pmc_files_detailed(ftp_comm_remote, filename_pattern=contains_file)
noncomm_xml = list_pmc_files_detailed(ftp_noncomm_remote, filename_pattern=contains_file)
misc_xml = list_pmc_files_detailed(ftp_misc_remote, filename_pattern=contains_file)
```

```python
comm_size = sum([d['size_gb'] for d in comm_xml])
noncomm_size = sum([d['size_gb'] for d in noncomm_xml])
misc_size = sum([d['size_gb'] for d in misc_xml])

print(comm_size, '\n', noncomm_size, '\n', misc_size)
```

```python
comm_meta = load_and_concatenate_csvs(comm_csv)
comm_meta.to_parquet('../assets/pmid/COMM_LICENSE.parquet')
```

```python
noncomm_meta = load_and_concatenate_csvs(noncomm_csv, cache_to_disk=False)
noncomm_meta.to_parquet('../assets/pmid/NONCOMM_LICENSE.parquet')
```

```python
misc_meta = load_and_concatenate_csvs(misc_csv, cache_to_disk=False)
misc_meta.to_parquet('../assets/pmid/MISC_LICENSE.parquet')
```

```python
comm_meta = pd.read_parquet('../assets/pmid/COMM_LICENSE.parquet')
```
